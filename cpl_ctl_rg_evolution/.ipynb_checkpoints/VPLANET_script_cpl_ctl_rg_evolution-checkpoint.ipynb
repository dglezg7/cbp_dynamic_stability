```python
import subprocess
import sys
import os
import shutil

"""
NOTE: 
Double check to see if your primary.in file is set to 
bEvolveRG   1 (Radius of Gyration evolution is on)
sTideModel   p2 (CPL model is on)

Make sure that the secondary.in also has
bEvolveRG   1

Finally, make sure that vpl.in has its system name set to
sSystemName   fleming2018_cpl_rg_on
"""

f2018 = "fleming2018"
file_names = [ "_cpl_rg_on", "_cpl_rg_off", "_ctl_rg_off", "_ctl_rg_on", "_cpl_rg_on"]
file_types = ["primary.in", "secondary.in"]
for file in range(len(file_names) - 1):
    subprocess.call(["vplanet", "vpl.in"]) #Runs the simulation, which produces .forward and .log files
    subprocess.call(["sed", "-i", "-e", 
                 's/' + f2018 + file_names[file] + "/" 
                     + f2018 + file_names[file + 1] + '/g', 
                 "vpl.in"]) #changes file name
    for i in range(len(file_types)): #Changes names and parameters on both primary.in and secondary.in
        subprocess.call(["sed", "-i", "-e", 
                     's/' + f2018 + file_names[file] + "/" 
                         + f2018 + file_names[file + 1] + '/g', 
                     file_types[i]]) #changes file name
        current_rg_boolean = "on" in file_names[file] #We'll assume on as the default state
        next_rg_boolean = "on" in file_names[file + 1]
        if current_rg_boolean != next_rg_boolean and current_rg_boolean == True:
            """If statement considers rg evolution as on and switches the state to off if the
            next file name is set to off."""
            subprocess.call(["sed", "-i", "-e", 
                             's/' + "bEvolveRG       1" + "/" 
                             + "bEvolveRG       0" + '/g', 
                             file_types[i]]) #Switches RG evolution from on to off
        elif current_rg_boolean != next_rg_boolean and current_rg_boolean == False:
            """If statement considers rg evolution as off and switches the state to on if the
            next file name is set to on"""
            subprocess.call(["sed", "-i", "-e", 
                             's/' + "bEvolveRG       0" + "/" 
                             + "bEvolveRG       1" + '/g', 
                             file_types[i]]) #Switches RG evoluation from off to on
        elif current_rg_boolean == next_rg_boolean:
            """No need to change anything if both the current and the next state are the same"""
            pass
        
        current_model_bool = "cpl" in file_names[file] #We'll assume the cpl model as the default model
        next_model_bool = "cpl" in file_names[file + 1]
        if current_model_bool != next_model_bool and current_model_bool == True:
            """If statement considers tide model as cpl and switches the state to ctl if the
            next file name is set to ctl."""
            subprocess.call(["sed", "-i", "-e", 
                             's/' + 'sTideModel      p2' 
                             + '/sTideModel      t8' + '/g', 
                             "primary.in"]) #Switches from the cpl to the ctl model
            subprocess.call(["sed", "-i", "-e", 
                             's/' + 'dTidalQ         1e6' 
                             + '/dTidalTau      3.17e-9' + '/g', 
                             file_types[i]]) #Converts TidalQ constant to TidalTau constant
        elif current_model_bool != next_model_bool and current_model_bool == False:
            """If statement considers tide model as ctl and switches the state to cpl if the
            next file name is set to cpl."""
            subprocess.call(["sed", "-i", "-e", 
                             's/' + 'sTideModel      t8' 
                             + '/sTideModel      p2' + '/g', 
                             "primary.in"]) #Switches from the ctl to the cpl model
            subprocess.call(["sed", "-i", "-e", 
                             's/' + 'dTidalTau      3.17e-9' 
                             + '/dTidalQ         1e6' + '/g', 
                             file_types[i]]) #Converts TidalQ constant to TidalTau constant
        elif current_model_bool == next_model_bool:
            """No need to change anything if both the current and the next tide models are the same"""
            pass
        
        f = open(file_types[i], 'r') #Reads the file
        print(f.read()) #Prints the whole file on the terminal
        f.close() #Closes the file
```
